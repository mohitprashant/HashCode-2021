```python
superstring='f'
```

```python
streets={}
intersections=[]


file = open(superstring+'.txt', 'r').readlines()

firstline = file[0]
firstline = firstline.split(sep=None)
D = int(firstline[0])
I = int(firstline[1])
S = int(firstline[2])
V = int(firstline[3])
F = int(firstline[4])

```

```python
for i in range(I):
    intersections.append([])

for i in range(1, S+1):
    start, end, name, length = file[i].split()
    start=int(start)
    end=int(end)
    length=int(length)
    
    streets[name]=[start, end, length]
    intersections[end].append([name, 0])

```

```python
for i in range(S+1, S+1+V):
    temp = file[i].split()
    pathLength = int(temp[0])
    l=temp[1:]
    
    for x in l:
        intersect=streets[x][1]
        streetlength=streets[x][2]
        for j in range(len(intersections[intersect])):
            if(x in intersections[intersect][j]):
                intersections[intersect][j][1]+=(1*(1/pathLength)*F)/streetlength
```

```python
for i in range(len(intersections)):
    normalizer=F
    
    for j in range(len(intersections[i])):
        if(normalizer>intersections[i][j][1] and intersections[i][j][1]!=0):
            normalizer=intersections[i][j][1]
    
    for j in range(len(intersections[i])):
        intersections[i][j][1]=intersections[i][j][1]/(normalizer)
        intersections[i][j][1]=int(intersections[i][j][1]//1)
```

```python
fileout=open(superstring+"latestsolution.txt", "w")

countmegazero=0

for x in intersections:
    countultrazero=0
    
    for j in x:
        if(j[1]==0):
            countultrazero+=1
            
    if(countultrazero==len(x)):
        countmegazero+=1


fileout.write(str(len(intersections)-countmegazero)+'\n')

for i in range(len(intersections)):
    intersectionindex=i
    
    countzero=0
    for x in intersections[i]:
        if(x[1]==0):
            countzero+=1
    
    intersectionlength=len(intersections[i])-countzero
    
    if(intersectionlength==0):
        continue
    
    fileout.write(str(intersectionindex)+"\n")
    fileout.write(str(intersectionlength)+"\n")
    
    for j in range(len(intersections[i])):
        goodthings=""
        goodthings+=intersections[i][j][0]+' '
        
        number=int((float(intersections[i][j][1])**0.35)//1)
        
        if(number==0):
            continue
            
        goodthings+=str(number)+'\n'
        
        
        fileout.write(goodthings)
        
fileout.close()
```
